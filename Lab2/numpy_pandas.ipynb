```python
# BTVN 01

#Cho ma trận bằng = [[99,99,99], 
#                    [99,99,99], 
#                    [99,99,99]]
# Hãy tưởng tượng ô 99 trống
# Giả sử 1 là x
# Giả sử 0 là o

# Lấy đầu vào từ phía x và phía o luân phiên
# Nếu đầu vào từ phía x là (0,0) thì 
# ma trận sẽ trở thành [[x,99,99], 
#                       [99,99,99], 
#                       [99,99,99]]


# Nếu đầu vào từ phía o cũng là (0,0) thì 
#  + Thông báo nhập lại
#  + Nếu không thì điền vào ma trận

# Thử thách: Nếu người đầu tiên có 3 phần tử liên tiếp thì dừng trò chơi


import numpy as np

# Khởi tạo ma trận
board = np.full((3, 3), 99, dtype=object)

def play_step(row, col, player):
    if board[row, col] == 99:
        board[row, col] = player
        print(f"Người chơi {player} đánh vào ({row}, {col})")
        print(board)
    else:
        print("Ô này có người đánh rồi!")



play_step(0, 0, 'x')  # Lượt 1: x đánh (0,0)
play_step(1, 1, 'o')  # Lượt 2: o đánh (1,1)
play_step(0, 1, 'x')  # Lượt 3: x đánh (0,1)
```

```python
# BTVN 04

# Tạo một mảng numpy có kích thước 150 x 5. Giả sử mảng này chứa 150 mẫu dữ liệu của sinh viên bao gồm chiều cao, cân nặng, tuổi, lương, điểm trung bình
# Tách mảng dựa trên 4 cột đầu tiên thành 1 biến có tên là x và cột cuối cùng thành y 

# Tách X thành X_train và X_test trong đó X_train chứa 70% dữ liệu
# Tách Y thành Y_train và Y_test trong đó Y_train chứa 70% dữ liệu
# Tạo 10 tập dữ liệu X_train không chồng chéo nhau


# Tạo mảng ngẫu nhiên 
data = np.random.rand(150, 5)

# 1. Tách mảng thành X (4 cột đầu) và y (cột cuối)
X = data[:, :4]
y = data[:, 4]

# 2. Tách X_train, X_test (70% - 30%)
split_idx = int(0.7 * len(X))

X_train = X[:split_idx]
X_test = X[split_idx:]

# 3. Tách Y_train, Y_test
y_train = y[:split_idx]
y_test = y[split_idx:]

print(f"Kích thước X_train: {X_train.shape}, X_test: {X_test.shape}")

# 4. Tạo 10 tập dữ liệu X_train không chồng chéo
X_train_subsets = np.array_split(X_train, 10)
print(f"Số lượng tập con: {len(X_train_subsets)}, mỗi tập có shape: {X_train_subsets[0].shape}")
```
