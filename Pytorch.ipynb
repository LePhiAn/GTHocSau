```python
#BTVN 
# Cho y = 5x^5 + 6x^3 - 3x + 1 
# Cho biết độ dốc của đa thức trên ở điểm nào
```

```python
import torch

# Khởi tạo x
x = torch.tensor(2.0, requires_grad=True)

# Định nghĩa hàm số
y = 5*x**5 + 6*x**3 - 3*x + 1

# Tính toán gradient
y.backward()

print(f"Độ dốc của hàm số tại x = 2 là: {x.grad}")
```

```python
#BTVN 02
#Tạo một tensor ban đầu có giá trị là 2
#Định nghĩa hàm số và tính gradient
#y =x^3 + 2x^2 + 5x + 1
# Hãy tính dy/dx tại giá trị của x
# Dùng phương pháp Gradient Descent với Learning rate alpha = 0.1 để cập nhật giá trị x trong vòng 10 vòng lặp
```

```python
x = torch.tensor(2.0, requires_grad=True)
learning_rate = 0.1

for epoch in range(10):
    y = x**3 + 2*x**2 + 5*x + 1
    # Tính gradient
    y.backward()
    # Cập nhật x
    with torch.no_grad():
        x -= learning_rate * x.grad

        x.grad.zero_()
        
    print(f"Vòng lặp {epoch+1}: x = {x.item():.4f}, y = {y.item():.4f}")
```

## BTVN 03
Tạo một tập dữ liệu giả lập với x là số giờ học (ngẫu nhiên từ 1 tới 10) và y là số điểm được tính theo công thức y = 3x + 5 + Noise
Với Noise là một giá tị ngẫu nhiên nhỏ

1. Khởi tạo tham số w và b ngẫu nhiên với requirers_grad = True
2. Tính MSE
3. Tính Gradient
4. Cập nhật tham số w và b bằng Gradient Descent với learning rate alpha = 0.01
5. Lập lại quá trình trên trong 100 vòng lập để xem sự hội tụ của mô hình

```python
import torch

# 1. Khởi tạo dữ liệu giả lập
torch.manual_seed(100)
X = torch.randint(1, 11, (100, 1)).float()
noise = torch.randn(100, 1) * 0.5
Y = 3 * X + 5 + noise

# Khởi tạo w và b ngẫu nhiên
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
alpha = 0.01

#5 Lặp 100 lần
for epoch in range(100):
    # Dự đoán
    y_pred = X * w + b
    
    #2 Tính MSE 
    loss = torch.mean((y_pred - Y)**2)
    
    #3 Tính Gradient
    loss.backward()
    
    #4 Cập nhật tham số
    with torch.no_grad():
        w -= alpha * w.grad
        b -= alpha * b.grad
        
        # Xóa gradient
        w.grad.zero_()
        b.grad.zero_()

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}, w = {w.item():.4f}, b = {b.item():.4f}")
```
